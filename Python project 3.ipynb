```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('diwali sales data.csv',encoding='unicode_escape')
df
```

```python
df.shape
```

```python
df.info()
```

```python
#REMOVING THE THE NULL COLUMNS

df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
#dropping the the null cells in amount column

df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
df.columns
```

```python
# GENDERS
```

```python
plt.figure(figsize=(4,4))
x1=sns.countplot(data=df,x='Gender',color='y')
x1.bar_label(x1.containers[0])
plt.show()
```

```python
gb=df.groupby(['Gender']).agg({'Amount':'sum'})
gb
```

```python
plt.figure(figsize=(4,4))
x2=sns.barplot(data=gb,x='Gender',y='Amount',color='g')

plt.show()
```

```python
#NOTE:FROM ABOVE WE SEE THAT MOST OF THE BUYERS ARE FEMALES AND THEY HAVE HIGH PURCHASING POWER THAN MEN.
```

```python
# AGE-GROUP
```

```python
x3=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in x3.containers:x3.bar_label(bars)
plt.show()
```

```python
gb1=df.groupby(['Age Group']).agg({'Amount':'sum'}).sort_values(by='Amount',ascending=True)
gb1
```

```python
plt.figure(figsize=(8,5))
x4=sns.barplot(data=gb1,x='Age Group',y='Amount')

plt.show()
```

```python
# FROM ABOVE WE SEE THAT MOST OF BUYS ARE ARE OF AGE GROUP 26-35 YEARS FEMALE
```

```python
#STATE
```

```python
# TOP 10 STATES BY ORDERS


gb2=df.groupby(['State']).agg({'Orders':'count'}).sort_values(by='Orders',ascending=False).head(10)
gb2
```

```python
plt.figure(figsize=(8,5))
x5=sns.barplot(data=gb2,x='State',y='Orders',color='g')
for bars in x5.containers:x5.bar_label(bars)
plt.xticks(rotation=60)
plt.show()
```

```python
#TOP 10 STATES BY AMOUNT

gb3=df.groupby(['State']).agg({'Amount':'sum'}).sort_values(by='Amount',ascending=False).head(10)
gb3
```

```python
plt.figure(figsize=(8,5))
x6=sns.barplot(data=gb3,x='State',y='Amount')

plt.xticks(rotation=60)
plt.show()
```

```python
#NOTE:FROM ABOVE WE SEE THAT MOST OF THE ORDERS AND PURCHASE ARE FROM UTTARPRADESH,MAHARASTRA AND KARNATAKA.
```

```python
#*******Maritial status*********
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df,x='Marital_Status',color='g')
for bars in ax.containers:ax.bar_label(bars)
plt.show()
```

```python
gb4=df.groupby(['Marital_Status']).agg({'Amount':'sum'}).sort_values(by='Amount',ascending=False)
gb4
```

```python
#  NOTES:most buys are married women
```

```python
# OCCUPATION
```

```python
plt.figure(figsize=(10,6))
x7=sns.countplot(data=df,x='Occupation',color='y')
for bars in x7.containers:x7.bar_label(bars)
plt.xticks(rotation=60)
plt.show()
```

```python
gb5=df.groupby(['Occupation']).agg({'Amount':'sum'}).sort_values(by='Amount',ascending=False)
gb5
```

```python
sns.barplot(data=gb5,x='Occupation',y='Amount',color='g')
plt.xticks(rotation=60)
plt.show()
```

```python
# NOTE: from above we see that most buyers are from IT sector,Healthcare and Aviation
```

```python
#PRODUCT CATEGORY
```

```python
plt.figure(figsize=(10,5))
x8=sns.countplot(data=df,x='Product_Category')
for bars in x8.containers:x8.bar_label(bars)
plt.xticks(rotation=75)
plt.show()
```

```python
gb6=df.groupby(['Product_Category']).agg({'Amount':'sum'}).sort_values(by='Amount',ascending=False)
gb6
```

```python
sns.barplot(data=gb6.head(10),x='Product_Category',y='Amount',color='y')
plt.xticks(rotation=75)
plt.show()
```

```python
#  NOTES: most sold products are food,clothing ,and electronics and gadgets
```
